## Part 1: Regression analysis of Gapminder data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn

data = pd.read_csv("https://raw.githubusercontent.com/jennybc/gapminder/master/data-raw/08_gap-every-five-years.tsv", sep='\t')
data.head()
```

#### Exercise 1: Make a scatter plot of life expectancy across time.

```python
year = data['year']
lifeExp = data['lifeExp']

plt.figure(figsize=(12,6))
#plt.scatter(year, lifeExp, marker='.')
sbn.scatterplot(year, lifeExp, data=data, marker='.').set(xticks=data.year.unique())
plt.title("Life Expectancy for Multiple Countries Across Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.show()
```

#### Question 1: Is there a general trend (e.g., increasing or decreasing) for life expectancy across time? Is this trend linear?
